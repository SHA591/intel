```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

# Load the dataset
data = pd.read_csv('/content/updated_drone_dataset_with_new_features.csv')

# Separate features into numeric and categorical
numeric_features = ['wind_speed', 'wind_angle', 'battery_voltage', 'battery_current', 'position_x', 'position_y']
categorical_features = ['Message_Integrity', 'Signal_Interference', 'Encryption_Status', 'Intrusion_Attempts']

# Pipeline for numeric features (impute and scale)
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Pipeline for categorical features (impute and one-hot encode)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine transformations
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Apply transformations to data
X_preprocessed = preprocessor.fit_transform(data)

# Split the data into train and test sets
X_train, X_test = train_test_split(X_preprocessed, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Initialize Isolation Forest
iso_forest = IsolationForest(n_estimators=200, max_samples='auto', contamination=0.1, random_state=42)

# Train the Isolation Forest on the training data
iso_forest.fit(X_train)

# Predict anomalies on the test set
y_pred = iso_forest.predict(X_test)

# Convert predictions: -1 (anomaly) to 1 and 1 (normal) to 0
y_pred = np.where(y_pred == -1, 1, 0)

# If ground truth (y_test) is available, you can evaluate the model
# Assuming 'y_test' contains ground truth where 1 indicates anomaly and 0 indicates normal

# Uncomment if you have the ground truth labels
# print(confusion_matrix(y_test, y_pred))
# print(classification_report(y_test, y_pred))
```

```python
# Sample visualization using the first two principal components
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
# Reduce dimensionality for visualization
pca = PCA(n_components=2)
X_test_pca = pca.fit_transform(X_test)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_test_pca[:, 0], y=X_test_pca[:, 1], hue=y_pred, palette=['green', 'red'])
plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Anomaly', labels=['Normal', 'Anomaly'])
plt.show()
```
